```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

import numpy as np
iris_data = pd.read_csv("iris.csv")

```

# 1 Univariate Analysis

## 1.1 Histogram

```python
sns.distplot( iris_data.loc[iris_data['species'] == 'setosa']['petal_length'] , color="skyblue", label="setosa")
sns.distplot( iris_data.loc[iris_data['species'] == 'versicolor']['petal_length'] , color="orange", label="versicolor")
sns.distplot( iris_data.loc[iris_data['species'] == 'virginica']['petal_length'] , color="green", label="virginica")
plt.legend()
plt.title('Histogram of various types of iris flower based on peta length')
```

```python
sns.distplot( iris_data.loc[iris_data['species'] == 'setosa']['petal_width'] , color="skyblue", label="setosa")
sns.distplot( iris_data.loc[iris_data['species'] == 'versicolor']['petal_width'] , color="orange", label="versicolor")
sns.distplot( iris_data.loc[iris_data['species'] == 'virginica']['petal_width'] , color="green", label="virginica")
plt.legend()
plt.title('Histogram of various types of iris flower based on petal width')
```

```python
sns.distplot( iris_data.loc[iris_data['species'] == 'setosa']['sepal_length'] , color="skyblue", label="setosa")
sns.distplot( iris_data.loc[iris_data['species'] == 'versicolor']['sepal_length'] , color="orange", label="versicolor")
sns.distplot( iris_data.loc[iris_data['species'] == 'virginica']['sepal_length'] , color="green", label="virginica")
plt.legend()
plt.title('Histogram of various types of iris flower based on sepal length')
```

```python
sns.distplot( iris_data.loc[iris_data['species'] == 'setosa']['sepal_width'] , color="skyblue", label="setosa")
sns.distplot( iris_data.loc[iris_data['species'] == 'versicolor']['sepal_width'] , color="orange", label="versicolor")
sns.distplot( iris_data.loc[iris_data['species'] == 'virginica']['sepal_width'] , color="green", label="virginica")
plt.legend()
plt.title('Histogram of various types of iris flower based on sepal width')
```

## 1.2 Box plot

```python
sns.boxplot(x='species',y='petal_length', data=iris_data)
plt.show()
```

```python
sns.boxplot(x='species',y='petal_width', data=iris_data)
plt.show()
```

## 1.3 Violin Plot

```python
sns.violinplot(x="species", y="petal_length", data=iris_data)
plt.show()
```

```python
sns.violinplot(x="species", y="petal_width", data=iris_data)
plt.show()
```

# 2 Bivariate Analysis

## 2.1 2-D Scatter Plot

```python
sns.scatterplot(data=iris_data,x='sepal_length', y='sepal_width',hue='species') ;
plt.title('scatter plot based on sepal length')
```

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(iris_data, hue="species", size=5);
plt.title('Pair plot')
plt.show()
# NOTE: the diagnol elements are PDFs for each feature. PDFs are expalined below.
```
